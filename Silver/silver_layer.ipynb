```sql
%sql
CREATE OR REPLACE TABLE university_cata.silver.student_activity AS
SELECT
    id_student,
    code_module,
    code_presentation,
    date,
    SUM(sum_click) AS total_clicks
FROM university_cata.bronze.studentVle
GROUP BY
    id_student,
    code_module,
    code_presentation,
    date;
```

```sql
%sql
CREATE OR REPLACE TABLE university_cata.silver.student_performance AS
SELECT
    sa.id_student,
    a.code_module,
    a.code_presentation,
    AVG(sa.score) AS avg_score,
    SUM(sa.score * a.weight) / SUM(a.weight) AS weighted_score

FROM university_cata.bronze.studentAssessment sa

JOIN university_cata.bronze.assessments a
ON sa.id_assessment = a.id_assessment

GROUP BY
    sa.id_student,
    a.code_module,
    a.code_presentation;
```

```sql
%sql
CREATE OR REPLACE TABLE university_cata.silver.student_performance AS
SELECT
    sa.id_student,
    a.code_module,
    a.code_presentation,

    AVG(sa.score) AS avg_score,

    CASE 
        WHEN SUM(a.weight) = 0 THEN NULL
        ELSE SUM(sa.score * a.weight) / SUM(a.weight)
    END AS weighted_score

FROM university_cata.bronze.studentAssessment sa

JOIN university_cata.bronze.assessments a
ON sa.id_assessment = a.id_assessment

GROUP BY
    sa.id_student,
    a.code_module,
    a.code_presentation;
```

```sql
%sql
CREATE OR REPLACE TABLE university_cata.silver.dim_students AS
SELECT DISTINCT
    id_student,
    code_module,
    code_presentation,
    gender,
    region,
    highest_education,
    imd_band,
    age_band,
    num_of_prev_attempts,
    studied_credits,
    disability,
    final_result
FROM university_cata.bronze.studentInfo;
```

```sql
%sql
create or replace table university_cata.silver.dim_module as
select
    code_module,
    code_presentation,
    module_presentation_length
from university_cata.bronze.courses
```

```sql
%sql
CREATE OR REPLACE TABLE university_cata.silver.student_registration AS
SELECT
    id_student,
    code_module,
    code_presentation,
    date_registration,
    date_unregistration
FROM university_cata.bronze.studentRegistration;
```

duplication checks 


```sql
%sql
SELECT
    id_student,
    code_module,
    code_presentation,
    COUNT(*) AS cnt
FROM university_cata.silver.dim_students
GROUP BY
    id_student,
    code_module,
    code_presentation
HAVING cnt > 1;
```
