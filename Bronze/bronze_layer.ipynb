```python
%sql
CREATE SCHEMA IF NOT EXISTS university_cata.bronze;
CREATE SCHEMA IF NOT EXISTS university_cata.silver;
CREATE SCHEMA IF NOT EXISTS university_cata.gold;
```

```python
files = [
    "assessments",
    "courses",
    "studentAssessment",
    "studentInfo",
    "studentRegistration",
    "studentVle",
    "vle"
]
```

```python
for file in files:
    df = spark.read.csv(
        f"/Volumes/university_cata/default/university_vol/{file}.csv",
        header=True,
        inferSchema=True
    )
    
    df.write.format("delta") \
        .mode("overwrite") \
        .saveAsTable(f"university_cata.bronze.{file}")
```
